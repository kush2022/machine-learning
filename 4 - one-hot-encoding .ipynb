### One-hot-encoding 
- is a technique used to represent categorical variables as binary vectors in machine learning. 
- in one-hot encoding, each category in a categorical variable is transformed into a binary vector that has a lenght equal to the number of unique categories. 
- The vector as a value of 1 for the category it represents and 0 for all other categories. 

```python
import pandas as pd 
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('homeprices.csv')
df
```

## using pandas to create dummy variables 


```python
dummies = pd.get_dummies(df.town, drop_first=True)
dummies
```

```python
merged = pd.concat([df, dummies], axis=1)
merged
```

```python
final = merged.drop('town', axis=1)
final
```

## using sklearn OneHotEncoder 

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
dfle = df
dfle.town = encoder.fit_transform(dfle.town)
dfle
```
