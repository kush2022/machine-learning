## Random Forest 
- is an ensemble learning method the combines multiple decision trees to create a more robust and accurate model.
- the main idea behind random forest is to create an ensemble of decision trees, where each tree is trained on a random subset of the data and a random subset of the features. During training each decision tree is built independently usng a technique called bagging.
- bagging involves randomly sampling the training data with replacement to create a multiple subsets, and then training each tree on a different subset. 

```python
import pandas as pd 
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
dir(digits)
```

```python
import matplotlib.pyplot as plt 
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
```

```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
X =  digits.data
y = digits.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

### Confusion matrix 
- is a table that visualizes the performance of a classification model by displaying the counts of true positive, true negative, false positive, and false negative made by the model 
- it is useful tool for eveluating the accuracy and effectiveness of a classification model 

```python
from sklearn.metrics import confusion_matrix
y_predicted = model.predict(X_test)
```

```python
import seaborn as sns 
```

```python
cm = confusion_matrix(y_test, y_predicted)
```

```python
cm
```

```python
sns.heatmap(cm, annot=True)
```
